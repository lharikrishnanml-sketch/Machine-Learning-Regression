```python
import pandas as pd # import pandas means we are importing pandas library into our python proram so we can use it and pd means its a nickname
dataset=pd.read_csv("Salary.csv")#reads a csv file called salary and stores in dataset variable
```

```python
dataset# prints dataset variable
```

```python
independent=dataset[["Years of experience"]]#selects the column years of experience and returns into a variable called independent
```

```python
independent# prints independent variable
```

```python
dependent=dataset[["Salary"]]#selects the column salary and returns into the variable called dependent
```

```python
dependent#prints the variable 
```

```python
from sklearn .model_selection import train_test_split# imports train_test_split from library sklearn to split into test and training data sets
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)#we are calling the function by using both independent and dependent by using test data percentage more 70% by storing them in 4 variables

```

```python
from sklearn.linear_model import LinearRegression# here we are calling linearregression class from sklearn library
regressor=LinearRegression()# we are calling LinearRegression function and storing in regresor variable
regressor.fit(X_train,Y_train)# trains ML model on training data
```

```python
LinearRegression() - #running the function LinearRegression()
```

```python
weight=regressor.coef_# gets slop/weight of linear regression model after training
weight#printing weight to see the value
```

```python
bais=regressor.intercept_#gets regressor intercept of linear regression model
bais#printing the bais to see the output
```

```python
y_predict=regressor.predict(X_test)# uses input to make predictions to the ouput how well the model is trained
```

```python
from sklearn.metrics import r2_score# uses metrics function from sklibrary to see howaccurate the mdodel predicts
r_score=r2_score(Y_test,y_predict)# calculates how well the model predicts
r_score#printing rscore
```

```python
import pickle # to save model by importing pickle
filename="finalized_model_linear.sav"# what filename to save as
```

```python
pickle.dump(regressor,open(filename,'wb'))#writing the file in the directory
```
